```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

v = CountVectorizer(ngram_range=(1,3))
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
def preprocess(text):
    doc = nlp(text)
    #     for token in text:
    pretex = [(token.lemma_) for token in doc if not (token.is_stop or token.is_punct)]
    return " ".join(pretex)   
```

```python
import pandas as pd
```

```python
df = pd.read_json("News_Category_Dataset_v3.json",lines = True)
```

```python
df.shape
```

```python
df.head(5)
```

```python
# for class imbalanced problem we have 4 techs. The simplest is undersampling
# we take the min sample and ignore the rest of the data 
min_sample = 1014
df[df.category == "EDUCATION"]
```

```python
# sampling data from politics i.e 1014 random samples 
# similar sampling(when we'll reload our screens) if we set a set random state as any number
df_politics = df[df.category == "POLITICS"].sample(min_sample,random_state=1687)
```

```python
df_women = df[df.category == "WOMEN"].sample(min_sample,random_state=1687)
df_crime = df[df.category == "CRIME"].sample(min_sample,random_state=1687)
df_parenting = df[df.category == "PARENTING"].sample(min_sample,random_state=1687)
df_arts = df[df.category == "ARTS"].sample(min_sample,random_state=1687)
```

```python
# concatenates df row by row
balanced_df = pd.concat([df_politics,df_women,df_crime,df_parenting,df_arts], axis = 0)

# balanced dataset
balanced_df.category.value_counts()
```

```python
# for mapping numerical values to categories
balanced_df["category_num"]=balanced_df["category"].map({
    "POLITICS":0,
    "WOMEN":1,
    "CRIME":2,
    "PARENTING":3,
"ARTS":4})
balanced_df
```

```python
from sklearn.model_selection import train_test_split

# splitting data set of balanced_df based on the category_num
X_train,X_test,y_train,y_test = train_test_split(balanced_df["short_description"],balanced_df.category_num,test_size=0.2,random_state=1687,
                                                stratify=balanced_df.category_num)
# equal no of samples from all the categories using stratify
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.value_counts()
```

```python
#  for bag of words ,model 
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

pipe = Pipeline([
    ("vectorizer_bow",CountVectorizer()),
    ("Multi_NB",MultinomialNB())
])

# training the model
pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print(classification_report(y_test,y_pred))

```

```python
# applying model with preprocessed text
balanced_df["preprocessed_text"]=balanced_df["short_description"].apply(preprocess)
```

```python
balanced_df.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(balanced_df["preprocessed_text"],balanced_df.category_num,test_size=0.2,random_state=1687,
                                                stratify=balanced_df.category_num)
```

```python
 
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

pipe = Pipeline([
    ("vectorizer_bow",CountVectorizer()),
    ("Multi_NB",MultinomialNB())
])


pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

# improved precision with preprocessed text
print(classification_report(y_test,y_pred))
```
